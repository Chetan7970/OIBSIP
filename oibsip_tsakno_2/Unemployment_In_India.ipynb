```python
df.columns=df.columns.str.replace(" ","")
```

```python
df["Date"]=df.Date.astype("datetime64")
```

```python
df['month_int'] =  df['Date'].dt.month
df['month'] = df['Date'].dt.strftime('%b')
df.head(20)
```

```python
grp=df.groupby(['month'])[['EstimatedUnemploymentRate(%)', "EstimatedEmployed", "EstimatedLabourParticipationRate(%)"]].mean()
grp=pd.DataFrame(grp).reset_index()
```

```python
grp.head()
```

```python
month = grp.month
estimated_unemployment_rate = grp["EstimatedUnemploymentRate(%)"]
estimated_labour_participation_rate = grp["EstimatedLabourParticipationRate(%)"]

fig = go.Figure()

fig.add_trace(go.Bar(x=month, y=estimated_unemployment_rate, name="Unemployment Rate", marker=dict(color='black')))
fig.add_trace(go.Bar(x = month, y = estimated_labour_participation_rate, name= "Labour Participation Rate",marker=dict(color="grey")))

fig.update_layout(title="Uneployment Rate and Labour Participation Rate")
fig.update_xaxes(categoryorder="array", categoryarray=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"])
fig.show()
```

```python
fig=px.bar(grp,x='month',y='EstimatedEmployed',color='month', category_orders = {"Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"]}, 
    color_discrete_map={   
        "Jan": "red",
        "Feb": "blue",
        "Mar": "green",
        "Apr": "orange",
        "May": "purple",
        "Jun": "yellow",
        "Jul": "cyan",
        "Aug": "magenta",
        "Sep": "brown",
        "Oct": "pink" },
             title='Estimated Employed People from Jan 2020 to Oct 2020')
fig.show()
```

```python
before_lockdown=df[(df["month_int"]>=1) & (df['month_int']<4)]
after_lockdown=df[(df['month_int']>=4) & (df['month_int']<=11)]
```

```python
af_lockdown=after_lockdown.groupby('Region')['EstimatedUnemploymentRate(%)'].mean().reset_index()
lockdown = before_lockdown.groupby('Region')['EstimatedUnemploymentRate(%)'].mean().reset_index()
lockdown['Unemployment Rate After Lockdown']=af_lockdown['EstimatedUnemploymentRate(%)']
lockdown.columns=['Region','Unemployment Rate Before Lockdown','Unemployment Rate After Lockdown']
lockdown["Rate of change in unemployment"]=round(lockdown["Unemployment Rate Before Lockdown"]-lockdown['Unemployment Rate After Lockdown']/lockdown['Unemployment Rate After Lockdown'],2)
```

```python
lockdown.head()
```

```python
fig=px.bar(lockdown,x="Region",y="Unemployment Rate Before Lockdown",color='Unemployment Rate Before Lockdown')
fig.update_layout(xaxis=({'categoryorder':'total ascending'}))
fig.show()
```

```python
fig=px.bar(lockdown,x="Region",y="Unemployment Rate After Lockdown",color='Unemployment Rate After Lockdown')
fig.update_layout(xaxis=({'categoryorder':'total ascending'}))
fig.show()
```

```python
fig=px.bar(lockdown,x="Region",y="Rate of change in unemployment",color='Rate of change in unemployment')
fig.update_layout(xaxis=({'categoryorder':'total ascending'}))
fig.show()
```
